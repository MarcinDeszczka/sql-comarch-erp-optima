##Zadanie 2.

**W bazie danych mamy listę produktów, ale handlowiec nie wie, ile one kosztują, bo nazwy są w jednym rejestrze, a kwoty w drugim. Musimy mu dostarczyć listę, gdzie obok nazwy towaru pojawi się jego cena netto, używając skrótów, aby kod był czytelny dla innych programistów. Ceny zaprezetuj w kolejności malejącej.**


```python
SELECT C.TwC_Wartosc, T.Twr_Nazwa
FROM CDN.TwrCeny AS C INNER JOIN CDN.Towary AS T
ON C.TwC_TwrId = T.Twr_TwrId
ORDER BY C.TwC_Wartosc DESC
```

##Zadanie 3. 

**W systemie Comarch ERP Optima każdy towar może posiadać wiele zdefiniowanych poziomów cenowych (kategorii). Aby ocenić ogólną "wagę" cenową asortymentu w bazie, konieczne było przygotowanie zestawienia sumującego wartości wszystkich kategorii cenowych dla każdego indeksu towarowego.**

```python

SELECT T.Twr_Kod, SUM(C.TwC_Wartosc) AS Suma_Cen
FROM CDN.Towary AS T INNER JOIN CDN.TwrCeny AS C 
ON T.Twr_TwrID = C.TwC_TwrID
GROUP BY T.Twr_Kod
ORDER BY Suma_Cen DESC
```

##Zadanie 4. "Raport Towarów Premium"

**Wyobraź sobie, że szef chce wiedzieć, które towary mają średnią cenę kategorii wyższą niż 150 PLN, ale interesują go tylko towary, które mają przypisane więcej niż 2 kategorie cenowe (żeby odsiać błędy lub pojedyncze wpisy).**

```python
SELECT T.Twr_Kod, AVG(C.TwC_Wartosc) AS Srednia_Cen
FROM CDN.Towary AS T INNER JOIN CDN.TwrCeny AS C 
ON T.Twr_TwrID = C.TwC_TwrID
GROUP BY T.Twr_Kod
HAVING COUNT(*) > 2 AND AVG(C.TwC_Wartosc) > 150
ORDER BY Srednia_Cen DESC
```

##Zadanie 5.

**Firma posiada dwie ceny: Hurtową (Numer 1) i Detaliczną (Numer 2). Szef podejrzewa, że wdał się błąd ludzki i dla niektórych towarów cena hurtowa jest... wyższa niż detaliczna. Musisz wyłapać te błędy.**

```python
SELECT T.Twr_Kod, T.Twr_Nazwa, C.TwC_Wartosc AS 'Cena hurtowa', C2.TwC_Wartosc AS 'Cena detaliczna'  
FROM CDN.Towary AS T
INNER JOIN CDN.TwrCeny AS C ON T.Twr_TwrID = C.TwC_TwrID AND C.TwC_TwCNumer = 1
INNER JOIN CDN.TwrCeny AS C2 ON T.Twr_TwrID = C2.TwC_TwrID AND C2.TwC_TwCNumer = 2
WHERE C.TwC_Wartosc > C2.TwC_Wartosc 
```
